```python
# Criando uma matriz numpy
import numpy as np
#Criando uma matriz de zeros utilizando np.zeros
a = np.zeros((5,5)) 
print(a)
print('indexação numpy')

print(a[0,0])
```

```python
#criando numeros aleatorios e mostrando a dimensão da matriz)
b = np.random.rand(5,5)
print(b)
print('dimensão da matriz')
print(b.shape)
```

```python
# Criando Matriz de valores aleatórios entre uma faixa estabelecida
vmin = 1 
vmax = 100

c = vmin+(vmax-vmin)*b
print(c)
```

```python
# Plotar matriz
import matplotlib.pyplot as plt
plt.plot(c.transpose())
```

```python
#Plotando apenas a primeira linha
x = np.array(range(0,np.size(c[0,:])))
print(x)
plt.xticks(x)
plt.plot(x,c[0,:])
```

```python
#criando um grafico de barras e exportando

fig, ax = plt.subplots(3,1)
ax[0].bar(x,c[:,0])
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,1])

import os 
rootFolder = os.getcwd()
print(rootFolder)
figFolder = rootFolder.replace('\\','/')+'/figures/'
print(figFolder)
fig.savefig('Figure01.png')
```
