```python
import numpy as np

# Dados originais
anos = np.arange(2007, 2013)
producao = np.array([176, 169, 190, 200, 195, 200])

# Calculando média e desvio padrão
media = np.mean(producao)
desvio_padrao = np.std(producao)

# Gerando números aleatórios para os anos de 2013 a 2050
anos_estendidos = np.arange(2013, 2051)
producao_estendida = np.random.normal(media, desvio_padrao, len(anos_estendidos))

# Exibindo os números aleatórios gerados
print("Produção de celulose gerada aleatoriamente até 2050:")
for ano, valor in zip(anos_estendidos, producao_estendida):
    print(f"Ano {ano}: {valor:.2f} mil toneladas")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados originais
anos = np.arange(2007, 2013)
producao = np.array([176, 169, 190, 200, 195, 200])

# Calculando média e desvio padrão
media = np.mean(producao)
desvio_padrao = np.std(producao)

# Gerando números aleatórios para os anos de 2013 a 2050
anos_estendidos = np.arange(2013, 2051)
producao_estendida = np.random.normal(media, desvio_padrao, len(anos_estendidos))

# Plotagem do gráfico
plt.figure(figsize=(10, 6))
plt.plot(anos, producao, marker='o', linestyle='-', label='Dados Originais')
plt.plot(anos_estendidos, producao_estendida, marker='o', linestyle='-', label='Dados Estendidos')
plt.title('Produção Anual de Celulose (Mil Toneladas)')
plt.xlabel('Ano')
plt.ylabel('Produção (Mil Toneladas)')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Exibição do gráfico
plt.show()
```
